```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
```

```python
tech_data = {
    "Track": [
        "Artificial Intelligence",
        "Web Development",
        "Data Science",
        "Cyber Security",
        "Cloud Computing"
    ],

    "Skills": [
        "Python Machine Learning Deep Learning TensorFlow PyTorch AI",
        "HTML CSS JavaScript React Node Web Frontend Backend",
        "Python Pandas NumPy Data Analysis Visualization Statistics",
        "Ethical Hacking Network Security Linux Penetration Testing",
        "AWS Azure Docker Kubernetes DevOps Cloud"
    ]
}

df = pd.DataFrame(tech_data)

print(df)
```

```python
vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(df["Skills"])

print(tfidf_matrix.shape)
```

```python
user_interest = input("Enter your interest: ")
```

```python
user_vector = vectorizer.transform([user_interest])
```

```python
similarity_scores = cosine_similarity(
    user_vector,
    tfidf_matrix
)

print(similarity_scores)
```

```python
best_match_index = similarity_scores.argmax()

print("Best Match Index:", best_match_index)
```

```python
recommended_track = df.iloc[best_match_index]["Track"]

print("\nRecommended Track:")
print(recommended_track)
```

```python
score = similarity_scores[0][best_match_index]

print("Similarity Score:", round(score * 100, 2), "%")
```

```python
scores = similarity_scores.flatten()

top_indices = scores.argsort()[-3:][::-1]

print("\nTop 3 Recommendations:\n")

for i in top_indices:
    print(
        f"{df.iloc[i]['Track']} "
        f"({scores[i]*100:.2f}%)"
    )
```
